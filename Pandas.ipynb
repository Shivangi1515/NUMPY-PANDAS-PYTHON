```python
#Melt and Pivot

df=pd.DataFrame({
    "country":["USA","USA","India","India"],
    "year":[2020,2021,2020,2021],
    "sales":[100,120,90,110],
    "profit":[20,25,18,22],
    "tax":[10,20,30,40]
})

melted_df=df.melt(  #wide to long format
    id_vars=["country","year"], #columns we want to keep
    value_vars=["sales","profit","tax"], #columns we want to melt
    var_name="metrics", #column name we want to give
    value_name="value" #column value we want  to store
)

melted_df.pivot(  #long to wide format
    index=["country","year"],
    columns="metrics",
    values="value"
)
```

```python
#Merging & Joining Data

df_customers=pd.DataFrame({
    "customer_id":[1,2,3,4],
    "name":["Adam","Bob","Charlie","Dave"]
})

df_orders=pd.DataFrame({
    "order_id":[101,102,103,104],
    "customer_id":[2,1,4,5],
    "amount":[250,120,300,180]
})

pd.merge(df_customers,df_orders,on="customer_id") #inner join

pd.merge(df_customers,df_orders,on="customer_id",how="left") #left join

pd.merge(df_customers,df_orders,on="customer_id" ,how="right") #right join

pd.merge(df_customers,df_orders,on="customer_id",how="outer") #outer join
```

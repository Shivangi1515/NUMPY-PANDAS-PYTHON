```python
#Cleaning the data-->Handling the missing values

df=pd.read_csv("raw_data.csv")
df.isnull()  #also reperesented by isna()

df.isnull().sum()

df.dropna() #drop rows-->if any row is having the null value it will drop that row
df.dropna(axis=1) #drop columns

df.fillna(0) #fill the null values with that value

age_mean=df["age"].fillna(df["age"].mean())

df["age"].fillna(age_mean)

age_mean=df["age"].mean()
cleaned_data=df.copy()
cleaned_data["age"]=cleaned_data["age"].fillna(age_mean)

cleaned_data

df.ffill() #forward fill-->fill the missing value from upward with last encountered value

df.bfill() #backward fill->fill the missing value from backward with last encountered value

```
